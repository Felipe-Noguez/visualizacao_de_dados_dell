# 1. Importando as bibliotecas e carregando os dados para um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

Exibição de 8 elementos aleatórios

```python
shampoo_sales = pd.read_csv('dadosoficina_3_3.txt')
shampoo_sales.sample(8)
```

# 2. Para o segundo e último passo, crie uma visualização que mostre a série temporal das vendas ao longo dos meses. Você pode, por exemplo, se perguntar se existe alguma tendência ou alguma periodicidade, OK?

```python
plt.figure(figsize=(19,6))
plt.title('Variação de vendas de Shampoo')
sns.lineplot(x='Month', y='Sales', data = shampoo_sales)
plt.show()
```

Nesta visualização do gráfico, podemos ver um pico maior a cada 2 ou 3 meses, a procura pelo produto foi crescente e os maiores picos de cada ano foram 1-11, 2-10 e 3-09.
