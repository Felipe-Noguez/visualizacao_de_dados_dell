<h2>1. Importando bibliotecas.<h2/>

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

<h2>Carregando dataframe<h2/>

```python
filmes_json = pd.read_json('movies.json')
filmes_json.head()
```

<h2>2. Gr√°fico com o custo total de cada filme<h2/>

```python
plt.figure(figsize=(8, 30))
sns.set(style = 'darkgrid')
plt.title('Custo total por filme!')
sns.barplot(y = 'Film', x = 'Budget_M', color = 'c', ci = None, data = filmes_json)
plt.show()
```

<h2>Exibindo filme com maior e menor custo<h2/>

```python
plt.figure(figsize=(8, 8))
sns.set(style = 'darkgrid')
plt.title('Filme com maior e menor custo.')
maior_custo = filmes_json.sort_values(by='Budget_M', ascending = False).head(1)
menor_custo = filmes_json.sort_values(by='Budget_M', ascending = True).head(1)
df_maior_custo = pd.concat([menor_custo, maior_custo])
sns.barplot(x = 'Film', y = 'Budget_M', color = 'c', ci = None, data = df_maior_custo);
```

<h2>3. Gr√°fico com bilheteria total de todos os filmes<h2/>

```python
plt.figure(figsize=(8, 30))
sns.set(style = 'darkgrid')
plt.title('Bilheteria total por filme!')
sns.barplot(y = 'Film', x = 'Worldwide_Gross_M', color = 'c', ci = None, data = filmes_json)
plt.show()
```

<h2>Exibindo filmes com maior e menor bilheteria<h2/>

```python
menor = filmes_json.sort_values(by='Worldwide_Gross_M', ascending = True).head(1)

maior = filmes_json.sort_values(by='Worldwide_Gross_M', ascending = False).head(1)
df_filmes = pd.concat([menor, maior])
sns.barplot(x = 'Film', y = 'Worldwide_Gross_M', color = 'c', estimator = sum, ci = None, data = df_filmes);
```

<h2>4. Gr√°fico com quantidade de filmes por ano<h2/>

```python
plt.figure(figsize=(20, 10))
sns.set(style = 'darkgrid')
plt.title('Bilheteria total por filme!')
#sns.countplot(x = 'Year', hue = 'Film',  data = filmes_json)
sns.countplot(x= 'Year', hue = 'Genre', palette = 'PuOr', edgecolor = '0', data = filmes_json)
plt.show()
```

<h2>Quantidade de filmes de a√ß√£o e filmes de anima√ß√£o feitas por ano<h2/>

```python
plt.figure(figsize=(20, 10))
sns.set(style = 'darkgrid')
plt.title('Quantidade de filmes de a√ß√£o e filmes de anima√ß√£o feitas por ano.')
data_action_animation = filmes_json.query('Genre == "Action" | Genre == "Animation"')
data_action_animation
sns.countplot(x = 'Year',hue = 'Genre', palette = 'colorblind', edgecolor = '0.4', data = data_action_animation);
```

<h2>5. Gr√°fico com m√©dia de or√ßamento por g√™nero<h2/>

```python
plt.figure(figsize=(20, 10))
sns.set(style = 'darkgrid')
plt.title('M√©dia de or√ßamento por g√™nero!')
sns.barplot(x='Genre', y='Budget_M', color='c', ci = 'sd', data=filmes_json)
plt.show()
```

<h2>M√©dia de or√ßamento entre A√ß√£o e Fantasia por ano<h2/>

```python
d10 = filmes_json.query('Genre == "Action" | Genre == "Animation"')
```

```python
plt.figure(figsize=(20, 10))
sns.set(style = 'whitegrid')
plt.title('M√©dia de or√ßamento entre A√ß√£o e Fantasia por ano.')
sns.barplot(x='Year', y='Budget_M', hue = 'Genre', palette="Set2", ci = 'sd', data=d10)
plt.show()
```
