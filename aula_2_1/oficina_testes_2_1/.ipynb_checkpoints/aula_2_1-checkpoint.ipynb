# Importando bibliotecas e carregando o arquivo com os dados (web e .json)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#atribuindo arquivo JSBON web à variável
filmes_json = 'https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json'
#lendo e atribuindo arquivo JSON em variavel
filmes_web = pd.read_json('https://raw.githubusercontent.com/emanueles/datavis-course/master/assets/files/observable/movies.json')
#Atribuindo arquivo json à variável
filmes_json_1 = 'movies.json'
#Atribuindo/lendo arquivo JSON da variavel
filmes = pd.read_json(filmes_json_1)
```

# Exibindo os dados das variáveis com os dados

```python
#filmes.head()
filmes_web
```

# Describe() para uma analise preliminar

```python
filmes_web.describe()
```

# Agrupando os dados com o método groupby()

```python
contagens = filmes.groupby(by='Genre').count()
contagens
```

# Gerando gráfico à partir do agrupamento
Agora você já possui os dados necessários para gerar a visualização utilizando o matplotlib. Basta selecionar o index e qualquer coluna do DataFrame contagens. Confira a seguir: 

```python
plt.figure(figsize=(9, 4))
plt.bar(contagens.index, contagens['Worldwide_Gross_M'], color = 'red')

plt.show()
```

# Gráficos de barras com Seaborn

```python
plt.figure(figsize=(9, 4))
sns.set(style='whitegrid')
sns.countplot(x='Genre', data=filmes)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.set(style='darkgrid')
plt.title('Quantidade de filmes opr gênero!')
sns.countplot(x='Genre', color='r', data=filmes)
plt.show()
```

```python
plt.title('Quantidade de filmes por gênero!')
sns.countplot(y='Genre', color='r', data=filmes)
plt.show()
```

```python
plt.title('Quantidade de filmes por ano!')
sns.countplot(y='Year', color='r', data=filmes)
plt.show()
```

# Calculando valores com SEABORN

```python
plt.figure(figsize=(10, 5))
plt.title('Bilheteria média por gênero!')
sns.barplot(x='Genre', y='Worldwide_Gross_M', color='b', ci = 'sd', data=filmes)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Bilheteria total por gênero!')
sns.barplot(x='Genre', y='Worldwide_Gross_M', color='orange', estimator = sum, ci = None, data=filmes)
plt.show()
```

# Tópico 2 – Gráficos de Barras Agrupadas
As linhas 1 e 2 configuram o tamanho e o título da figura, como já feito anteriormente. Na linha 3 é chamada a função countplot() de maneira muito similar ao que já foi feito, passando o valor ‘Year’ para o parâmetro x. Ou seja, os anos serão mapeados na posição do eixo x. A novidade é o parâmetro hue, que recebe a coluna ‘Genre’, significando que o gênero dos filmes será mapeado nas cores das barras. 

```python
plt.figure(figsize=(15,8))
plt.title('Filmes por gênero em cada ano.')
sns.countplot(x = 'Year', hue = 'Genre', data = filmes)
plt.show()
```

O parâmetro hue também pode ser utilizado na função barplot(). Então para criar essa visualização é análogo ao que foi feito com a função countplot(). Confira no exemplo: 

```python
plt.figure(figsize = (15, 8))
plt.title('Média de bilheteria por gênero em cada ano.')
sns.barplot(x = 'Year', y = 'Worldwide_Gross_M', ci = None, hue = 'Genre', data = filmes)
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.title('Média de bilheteria por gênero em cada ano')
sns.countplot(x = 'Year',hue = 'Genre', palette = 'pastel', edgecolor = '0.4', data = filmes)
plt.show()
```

# Salvando gráficos criados!

```python
plt.figure(figsize=(15,8))
plt.title('Média de bilheteria por gênero em cada ano')
sns.countplot(x="Year", hue= 'Genre', palette = "pastel", edgecolor="0", data=filmes)
plt.savefig('Média de bilheteria por gênero em cada ano.png')
```
