```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

Carregando os dados e exibindo primeras 5 ocorrências

```python
df = pd.read_csv('dadosoficina05.txt')
df.head()
```

```python
df['area'].describe()
```

```python
df.info()
```

# Criando o primeiro histograma

```python
plt.figure(figsize=(10,5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribuição de áreas dos imóveis')
plt.hist(df['area'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.style.use('seaborn-darkgrid')
plt.title('Distribuição de áreas dos imóveis')
plt.hist(df['area'],bins=25)
plt.show()
```

# Tópico 2 – Gráficos de densidade com seaborn

```python
plt.figure(figsize=(10, 5))
plt.title('Distribuição de áreas dos imóveis')
sns.distplot(df['area'])
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.title('Distribuição de áreas dos imóveis')
sns.kdeplot(df['area'], shade=True)
plt.show()
```
