# 1.1 Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

Carregando base de dados

```python
df = pd.read_csv('dadosoficina05.txt')
df.sample(15)
```

# 2. Crie uma visualização que mostre a distribuição dos valores de aluguel, ‘rent amount'  dos imóveis. Pode utilizar o gráfico que você julgar mais adequado. Qual faixa de valores de aluguel é mais comum? 

```python
plt.figure(figsize=(10, 4))
plt.title('Distribuição de valores de alugueis de imóveis')
sns.distplot(df['rent amount (R$)'])
plt.show()
```

A Faixa entre de 10.000 mil reais e 25.000 mil reais são as mais comuns.

# 3. Crie uma visualização que mostre a distribuição dos valores de seguro contra incêndio, 'fire insurance', dos imóveis. Existem valores incomuns? 

```python
plt.figure(figsize=(10, 4))
plt. title('Distribuição de valores de seguro contra incêndio')
sns.distplot(df['fire insurance (R$)'])
plt.show()
```

Valores mais raros estão entre 100 reais e 305 reais.
