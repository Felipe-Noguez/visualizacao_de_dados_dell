```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#sns.set(style='darkgrid')
```

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.isnull().sum()
```

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.stripplot(x="class", y='age', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.swarmplot(x="class", y='age', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('Idade de passageiros em cada classe')
sns.swarmplot(x="class", y='age', hue='sex', data=titanic)
plt.show()
```

```python
sns.catplot(x="pclass", y='age', hue='sex', col='survived', kind='swarm', data=titanic);
```

# Tópico 2 – Gráficos de caixa

```python
titanic.describe()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros de cada classe')
sns.boxplot(x='class', y='age', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros de cada classe por sexo')
sns.boxplot(x='class', y='age', hue='sex', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros de cada classe por sexo')
sns.boxenplot(x='class', y='age', hue='sex', data=titanic)
plt.show()
```

Idade, sexo, classe de passageiros que sobreviveram ou não: gráfico de caixa em grade.
BOXEN

```python
sns.catplot(x='class', y='age', hue='sex', col='survived', kind='boxen', data=titanic);
```

BOX

```python
sns.catplot(x='class', y='age', hue='sex', col='survived', kind='box', data=titanic);
```

# Tópico 3 – Gráficos de violino

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passagerios em cada classe')
sns.violinplot(x='class', y='age', data=titanic)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passagerios em cada classe')
sns.violinplot(x='class', y='age', hue='sex',data=titanic)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passageiros de cada classe (combinando Gráficos violino e de enxame')
sns.violinplot(x='class', y='age', data=titanic)
sns.swarmplot(x='class', y='age', color='black', size=3, data=titanic)
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Idade de passagerios em cada classe por sexo')
sns.violinplot(x='class', y='age', hue='sex', split=True,data=titanic)
plt.show()
```

```python
sns.catplot(x="class", y="age", hue = 'sex', col = 'survived', kind = 'violin', split = True, data=titanic);
```
