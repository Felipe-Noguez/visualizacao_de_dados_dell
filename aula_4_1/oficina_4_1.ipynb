# 1. Importando as bibliotecas e carregando os dados para um DataFrame.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

Carregando os dados da biblioteca Seaborn

```python
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
titanic.info()
```

# 2. Crie uma visualização que mostre a distribuição do valor da passagem (“fare”), de acordo com a cidade na qual o passageiro embarcou ("embark_town"). Você pode escolher a técnica que julgar mais adequada; 

```python
plt.figure(figsize=(10,7))
plt.title('Valor da passagem de acordo com a cidade de embarque')
sns.boxplot(x='embark_town', y='fare', data=titanic)
plt.show()
```
