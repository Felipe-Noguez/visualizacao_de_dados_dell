# Importando a biblioteca pandas e fazendo a leitura do arquivo

```python
import pandas as pd
df = pd.read_csv('drinks.txt')
```

# 1. Qual a média e desvio padrão do consumo de cerveja, licor, vinho e total de álcool?

```python
print('A média de consumo é: \n{}\n\nO desvio padrão é:\n{}'.format(df.mean(), df.std()))
```

# 2. Qual o consumo de licor na França?

```python
france = df[df['country'] == 'France']

print('O consumo de licor na França é: {}'.format(france['spirit_servings'].values))
```

# 3. Quais os 5 países nos quais se consome mais vinho?

```python
print('Os 5 paises que mais consomem vinho é:')
df.sort_values(by='wine_servings', ascending = False).head()
```

# 4. Quais os países nos quais o consumo de cerveja está acima da média?

```python
média = df['beer_servings'].mean()
print(f'A média de consumo de cerveja nos paíese é: {média:.2f}')
df[df['beer_servings'] > média]
```

# 5. Quais países nos quais o consumo total de álcool está acima do Brasil, em ordem do menor para o maior?
Abaixo temos os dados referentes aos países que tem um consumo de alcool maior que o Brasil

```python
br = df[df['country'] == 'Brazil']
consumobr = br['total_litres_of_pure_alcohol'].mean()
df_br = df[df['total_litres_of_pure_alcohol'] >= consumobr]
df_br.sort_values(by='total_litres_of_pure_alcohol', ascending = True)
```
