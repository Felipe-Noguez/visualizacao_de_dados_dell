# Tópico 1 – Distribuições bivariadas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

```python
iris = sns.load_dataset('iris')
iris.sample(15)
```

```python
sns.jointplot(x='sepal_length', y='petal_length', kind='reg', data=iris);
```

```python
sns.distplot(iris['sepal_length']) #teste
```

```python
sns.jointplot(x='sepal_length', y='petal_length', kind='hex', color='black', data=iris);
```

```python
sns.jointplot(x='sepal_length', y='petal_length', kind='kde', data=iris);
```
