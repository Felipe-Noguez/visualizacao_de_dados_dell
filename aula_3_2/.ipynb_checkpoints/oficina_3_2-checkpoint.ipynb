# 1. Importando as bibliotecas

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style='darkgrid')
```

Carregando a base de dados com 7 exemplos aleatórios.

```python
df = sns.load_dataset('mpg')
df.sample(7)
```

Armazenando consultas à variáveis que serão usadas nos gráficos.

```python
df_usa = df[df['origin'] == 'usa']
df_japan = df[df['origin'] == 'japan']
df_europe = df[df['origin'] == 'europe']
slice_72 = df.query('model_year <=72')
slice_75 = df.query('model_year >= 73 & model_year <=75')
slice_78 = df.query('model_year >= 76 & model_year <=78')
slice_82 = df.query('model_year > 78')
```

# 2. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível diferenciar o país de origem (‘origin’). 
# Sabendo disso, existe uma correlação entre a aceleração e o consumo? De que tipo? A correlação é a mesma para os diferentes países de origem?
Na visualização do gráfico com reta de regressão, podemos ver que os carros Americanos e Europeus daquela época, não existe uma relação entre aceleração e consumo de combustível (sem correlação), já nos carros japones temos uma baixa correlação positiva.



Exemplo de gráfico com reta de regressão.

```python
sns.lmplot(x='acceleration', y='mpg', hue = 'origin', data = df);
```

Para uma visualização mais clara, temos abaixo os gráficos para cada país.

```python
sns.relplot(x='acceleration', y='mpg', col = 'origin', data = df);
```

Exemplos de gráficos com reta de regressão para cada país para uma melhor visualização e análise.

```python
sns.lmplot(x='acceleration', y='mpg', col = 'origin', data = df_usa), 
sns.lmplot(x='acceleration', y='mpg', col = 'origin', data = df_japan),
sns.lmplot(x='acceleration', y='mpg', col = 'origin', data = df_europe);
```

# 3. Crie uma visualização que mostre a relação entre a aceleração (‘acceleration’) e o consumo de combustível (‘mpg’) dos veículos, em que seja possível visualizar o ano do modelo (model_year).
# Existe alguma relação entre as três variáveis?
Analisando os gráficos podemos ver que em alguns anos como por exemplo no ano de 1971, temos uma  baixa correlação positiva, mas, a maioria dos anos não existe uma relação


Exemplo de gráfico com muita informação, onde éxibido a correlação de consumo de combustível, aceleração e ano, não conseguimos ter uma visualização do padrão.

```python
sns.lmplot(x='acceleration', y='mpg', hue = 'model_year', data = df)
plt.show()
```

Gráficos com dados anuais com a relação entre aceleração e consumo de combustível por país.

```python
sns.relplot(x='acceleration', y='mpg', col = 'model_year', hue = 'origin', data = slice_72);
```

```python
sns.relplot(x='acceleration', y='mpg', col = 'model_year', hue = 'origin', data = slice_75);
```

```python
sns.relplot(x='acceleration', y='mpg', col = 'model_year', hue = 'origin', data = slice_78);
```

```python
sns.relplot(x='acceleration', y='mpg', col = 'model_year', hue = 'origin', data = slice_82);
```

Gráfico geral com a relação entre aceleração e consumo de combustível por ano.

```python
sns.relplot(x='acceleration', y='mpg', hue = 'model_year', data = df);
```
